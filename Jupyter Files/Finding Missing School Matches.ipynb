```python
frequency_dict = {}

# Step 3: Iterate through each row of the DataFrame
num_rows=school_data.shape[0]
counter=0
for index, row in school_data.iterrows():
    # Check if the 'c27_FINAL' column has an empty value (NaN)
    if pd.isna(row['c27_FINAL']):
        counter+=1
        # Get the value of 'c27_est' for this row
        c27_est_value = row['c27_est']

        # Update the frequency count in the dictionary
        frequency_dict[c27_est_value] = frequency_dict.get(c27_est_value, 0) + 1

# Step 4: Print the count
print(1-(counter/num_rows))
```

```python
sorted_frequency_dict = dict(sorted(frequency_dict.items(), key=lambda item: item[1], reverse=True))

# Step 5: Print the sorted frequency dictionary
print(sorted_frequency_dict)
```

```python
words_to_remove = ["universidad", "san", 'santa', 'No sabe', 'duoc', 'inacap', ]
sorted_frequency_dict = {key: value for key, value in sorted_frequency_dict.items() if isinstance(key, str) and not any(word in key.lower() for word in words_to_remove)}
# Step 6: Print the final frequency dictionary after deletion
for key, value in sorted_frequency_dict.items():
    print(f"{key}: {value}")
```
